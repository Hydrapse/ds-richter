```python
import pandas as pd

DATA_DIR = "data"

df7533 = pd.read_csv(DATA_DIR + '/submission.csv', index_col='building_id')
df7534 = pd.read_csv(DATA_DIR + '/submission_7534.csv', index_col='building_id')
df7540 = pd.read_csv(DATA_DIR + '/submission_7540.csv', index_col='building_id')
dfnew = pd.read_csv(DATA_DIR + '/submission_new.csv', index_col='building_id')
```

```python
diff = df7533[df7533["damage_grade"]!=df7534["damage_grade"]]
diff_new = dfnew[df7540["damage_grade"]!=dfnew["damage_grade"]]
diff_new =diff_new.join(df7540,how="left",lsuffix='_new', rsuffix='_7540')
diff = diff.join(df7534,how="left",lsuffix='_7533', rsuffix='_7534')
# diff_new = diff_new.join(df7540,how="left",lsuffix='_new', rsuffix='_7540')
# pd.concat([diff,diff_new],axis=1)
diff
```

```python
for index, row in diff.iterrows():
    new = dfnew.loc[index,"damage_grade"]
    data7533 = df7533.loc[index,"damage_grade"]
    data7534 = df7534.loc[index,"damage_grade"]
    if(new == data7533):
        dfnew.loc[index] = df7534.loc[index]

diff.join(dfnew,how="left")
```
